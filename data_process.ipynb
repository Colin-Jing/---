```python
import pandas as pd
from tqdm import tqdm

# 创建一个空的 DataFrame
train_df = pd.DataFrame(columns=['content', 'channelName'])
test_df = pd.DataFrame(columns=['content', 'channelName'])
train_df
test_df

# 导入数据
for i in range(10):
    df = pd.read_excel(r'C:\Users\Colin_Jing\competition\中国软件杯\1617241934831197.xlsx',sheet_name=i)
    print(df.keys())
    print(df.head())
    
    content_train_df = df.loc[0:(df.shape[0]-int(df.shape[0]/10)),['content','channelName']]
    title_train_df = df.loc[0:(df.shape[0]-int(df.shape[0]/10)),['title','channelName']]
    title_train_df.rename(columns={'title':'content'},inplace = True)
    tem_train_df = pd.concat([content_train_df,title_train_df], axis = 0,join='outer',ignore_index=True)
    train_df = pd.concat([train_df,tem_train_df],axis=0,join='outer',ignore_index=True)
    
    content_test_df = df.loc[(df.shape[0]-int(df.shape[0]/10)+1):df.shape[0],['content','channelName']]
    title_test_df = df.loc[(df.shape[0]-int(df.shape[0]/10)+1):df.shape[0],['title','channelName']]
    title_test_df.rename(columns={'title':'content'},inplace = True)
    tem_test_df = pd.concat([content_test_df,title_test_df], axis = 0,join='outer',ignore_index=True)
    test_df = pd.concat([test_df,tem_test_df],axis=0,join='outer',ignore_index=True)
# 生成训练接
train_df.to_excel(r'./data/train.xlsx',index=False)
# 生成测试集
test_df.to_excel(r'./data/test.xlsx',index=False)
```

```python
import pandas as pd
from tqdm import tqdm

# 创建一个空的 DataFrame
train_df = pd.DataFrame(columns=['title', 'channelName'])
test_df = pd.DataFrame(columns=['title', 'channelName'])

# 导入数据
for i in range(10):
    df = pd.read_excel(r'C:\Users\Colin_Jing\competition\中国软件杯\1617241934831197.xlsx',sheet_name=i)
    print(df.keys())
    print(df.head())
    
    title_train_df = df.loc[0:200,['title','channelName']]
    title_test_df = df.loc[df.shape[0]-200:df.shape[0],['title','channelName']]
    train_df = pd.concat([train_df,title_train_df],axis=0,join='outer',ignore_index=True)
    test_df = pd.concat([test_df,title_test_df],axis=0,join='outer',ignore_index=True)

# 生成训练集
train_df.to_excel(r'./官方数据/train.xlsx',index=False)
test_df.to_excel(r'./官方数据/test.xlsx',index=False)

#类别更新
tmp_file = pd.read_excel(r'./官方数据/train.xlsx')
tmp_file1 = pd.read_excel(r'./官方数据/test.xlsx')

for i in range(len(tmp_file['channelName'])):
    if tmp_file['channelName'][i] == '综合体育最新':
        tmp_file['channelName'][i] = '体育'
    elif tmp_file['channelName'][i] == '体育焦点':
        tmp_file['channelName'][i] = '娱乐'
tmp_file.to_excel(r'./官方数据/train.xlsx',index=False)

for i in range(len(tmp_file1['channelName'])):
    if tmp_file1['channelName'][i] == '综合体育最新':
        tmp_file1['channelName'][i] = '体育'
    elif tmp_file1['channelName'][i] == '体育焦点':
        tmp_file1['channelName'][i] = '娱乐'
tmp_file1.to_excel(r'./官方数据/test.xlsx',index=False)

# 标签更新
label_list = ['财经','房产','教育','科技','军事','汽车','体育','游戏','娱乐','其他']
tmp_file = pd.read_excel(r'./官方数据/train.xlsx')
tmp_file1 = pd.read_excel(r'./官方数据/test.xlsx')
#
for k in range(len(label_list)):
    for i in range(len(tmp_file['channelName'])):
        if tmp_file['channelName'][i] == label_list[k]:
            tmp_file['channelName'][i] = k
tmp_file.to_excel(r'./官方数据/train.xlsx', index=False)

for k in range(len(label_list)):
    for i in range(len(tmp_file1['channelName'])):
        if tmp_file1['channelName'][i] == label_list[k]:
            tmp_file1['channelName'][i] = k
tmp_file1.to_excel(r'./官方数据/test.xlsx', index=False)
```

```python
from openpyxl import load_workbook,Workbook
with open(r'C:\Users\Colin_Jing\competition\中国软件杯\text\THUCNews\data\test.txt', 'r', encoding='UTF-8') as f:
    wb = Workbook()    # 创建工作簿
    ws = wb.active   # 激活工作表
    print(ws)
    print('输出文件所有工作表名：\n', wb.sheetnames)
    ws['A1'] = 'content'
    ws['B1'] = 'channelName'

    for line in tqdm(f):
        lin = line.strip()
        if not lin:
            continue
        content, label = lin.split('\t')
        if label =='0':
            label = 0
        elif label == '1':
            label = 1
        elif label == '2':
            label = 0
        elif label == '3':
            label = 2
        elif label == '4':
            label = 3
        elif label == '5':
            label = 9
        elif label == '6':
            label = 4
        elif label == '7':
            label = 6
        elif label == '8':
            label = 7
        elif label == '9':
            label = 8
        ws.append([content,label])

    
    wb.save(r'./data/test.xlsx')
```
